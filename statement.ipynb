```python
import pandas as pd
```

```python
from datetime import date, timedelta
start_date = date(2023, 12, 8)
```

```python
data = dh.getStatements(start_date, start_date)

history = pd.DataFrame(data['data'])
history.drop(["dhanClientId", "orderId","isin","tradingSymbol" ,"exchangeOrderId", "exchangeTradeId", "productType", "orderType", 
              "stampDuty", "createTime", "updateTime"],axis=1, inplace=True)
history.sort_values(by=["exchangeTime"],inplace=True)
history_FNO = history[history["exchangeSegment"] == "NSE_FNO"]
history_EQ = history[history["exchangeSegment"] == "NSE_EQ"]
history_FNO
```

```python
# history_FNO.apply(process_row,axis=1)
history_FNO["Script"] = history_FNO["customSymbol"].apply(lambda x: x.split(" ")[0])
history_FNO
```

```python
bankNifty = history_FNO[history_FNO["Script"] == "BANKNIFTY"]
```

```python
raw_data = {
    "under_process":{},
    "completed":{}
}
def testing(data):
    print("====")
    if not raw_data["under_process"]:
        raw_data["under_process"] = {
            "transactionType": data["transactionType"],
            "customSymbol":data["customSymbol"]
        }
def process_row(data):
    print("--")
    data.apply(testing,axis=1)
bankNifty.groupby("securityId").apply(process_row)
print(raw_data)
```
